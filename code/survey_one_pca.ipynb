```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.impute import KNNImputer
from sklearn.decomposition import PCA
```

```python
clean_df = pd.read_csv('~/Documents/Data_Science/projects/mi_stigma_pca/data/survey_one_data_clean.csv', index_col = 0)
clean_df.head()
```

```python
#selecting final variables (drop isolated and embarassed adjectives)
features = clean_df.iloc[:, np.r_[17:34, 35:55, 56:59]]
features = features.drop(features.columns[[17, 15, 35, 37]], 1)
features.info()
```

```python
sum(features.isnull().any())
```

```python
imputer = KNNImputer(n_neighbors=3)
imputed = imputer.fit_transform(features)
df_imputed = pd.DataFrame(imputed, columns=features.columns)
```

```python
sum(df_imputed.isnull().any())
```

```python
#splitting into depression and schziophrenia specific datasets
df_dep = df_imputed.iloc[:, 0:18]
df_schiz = df_imputed.iloc[:, 18:]
```

```python
df_dep.info()
```

```python
df_schiz.info()
```

```python
#matched t-test to confirm syndromes are rated differently
from scipy import stats

np.random.seed(101)
stats.ttest_rel(df_dep,df_schiz) #all statistically significant
```

```python
def get_pca_loadings(data, num_comp):
    pca = PCA(n_components= num_comp)
    
    #fit_transform() to calculate the coordinate system
    principalComponents = pca.fit_transform(data)
    print(pca.explained_variance_ratio_)
    
    loadings = pca.components_.T * np.sqrt(pca.explained_variance_)
    loading_matrix = pd.DataFrame(loadings, index=data.columns)
    
    return loading_matrix
```

```python
dep_loading_matrix = get_pca_loadings(df_dep, 3)
dep_loading_matrix
```

#### PC1 loadings:
sincere, trustworthy, friendly, moral, understanding, intelligent, skilled, confident, creative, capable, desirable, successful 
**(competence)**
#### PC2 loadings:
unpredictable, dangerous, angry, violent
**(danger)**
#### PC3 loadings: 
helpful, happy
**(likeability/warmth)**
